```python
import allantools
import pylab as plt
import numpy as np
t = np.logspace(0, 3, 100)  # tau values from 1 to 1000
y = data["Reference_oscillator_frequenc"]/9192631770  # Generate some frequency data
r = 0.5  # sample rate in Hz of the input data
(t2, ad, ade, adn) = allantools.oadev(y, rate=r, data_type="freq", taus=t)  # Compute the overlapping ADEV
fig = plt.loglog(t2, ad) # Plot the results
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.show()
```

```python
import requests
import json
import time

# URL de la API
url = 'http://localhost:5000/api/send_datainstrument'  # Reemplaza localhost y el puerto según tu configuración

for index, row in data.iterrows():
    datos = row.to_dict()
    
    # Realiza la solicitud POST a la API
    response = requests.post(url, json=datos)
    
    # Verifica si la solicitud fue exitosa
    if response.status_code == 201:
        print("Datos enviados correctamente.")
    else:
        print("Error al enviar los datos:", response.text)
    
    # Agrega un retraso de 10 segundos
    time.sleep(1)
```
